Creation of an tfidf dataframe

```python

from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
nltk.download('stopwords')
english_stopwords = stopwords.words('english')
stemmer = PorterStemmer()

def stem_text(text):
    tokens = word_tokenize(text.lower())  
    stemmed_tokens = [stemmer.stem(word) for word in tokens if word.isalpha()]  
    return " ".join(stemmed_tokens)

train_df["news_content"] = train_df["news_content"].astype(str).apply(stem_text)
test_df["news_content"] = test_df["news_content"].astype(str).apply(stem_text)
tfidf_vectorizer = TfidfVectorizer(
    lowercase=True,
    ngram_range=(1, 1),
    max_features=3000,
    stop_words=english_stopwords
)


tfidf_matrix = tfidf_vectorizer.fit_transform(train_df["news_content"])
train_tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
selected_columns = train_df[["impact_level", "impact_length"]]
train_tfidf_df = pd.concat([selected_columns.reset_index(drop=True), train_tfidf_df.reset_index(drop=True)], axis=1)
tfidf_test_matrix = tfidf_vectorizer.transform(test_df["news_content"])
tfidf_test_df = pd.DataFrame(tfidf_test_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
selected_columns = test_df[["impact_level", "impact_length"]]
tfidf_test_df = pd.concat([selected_columns.reset_index(drop=True), tfidf_test_df.reset_index(drop=True)], axis=1)
#train_tfidf_df.to_csv("datasets/update_datasets/train_tfidf_level.csv", index=False)
#tfidf_test_df.to_csv("datasets/update_datasets/test_tfidf_level.csv", index=False)
```
