#### Translate Japanese texts to English

The Japanese-language dataset contains unique and valuable attributes. To make it more accessible for analysis in this thesis, we will translate the Japanese texts into English. The only attribute requiring translation is the ‘Text’ column, as all other attributes are either already in English or numerical. For this task, we will employ the Opus-MT model, the same model previously used for translating French to English.

```python
train_jp = pd.read_json('datasets/competition/ML-ESG-3_Trainset_Japanese.json')
test_jp = pd.read_json('datasets/competition/ML-ESG3_Testset_Japanese.json')
test_jp
```

```python
train_jp['Text'] = train_jp['Text'].apply(data_enrichment.translate_japanese)
test_jp['Text'] = test_jp['Text'].apply(data_enrichment.translate_japanese)
```

```python
test_jp
```

```python
train_jp.to_csv("datasets/update_datasets/other_languages/train_jp_to_en.csv", index=False)
test_jp.to_csv("datasets/update_datasets/other_languages/test_jp_to_en.csv", index=False)
```

```python
test_jp
```

The dataset will be cleaned to retain only the information necessary for the analysis. Specifically:

- Only entries relevant to ESG issues will be considered. Therefore, rows where the column "Relevancy" has the value "Relevant" will be kept.
- ESG issues are categorized into three main factors: Environment, Social, and Governance. Accordingly, new columns corresponding to these factors will be created, and the respective percentage values will be derived from the ESG_type attribute.
- The impact_type attribute is also important, as it relates directly to sentiment analysis. In this context, opportunity indicates that the news has a positive impact, whereas risk signifies a negative impact.

##### Relevancy

```python
train_jp_filtered = train_jp[train_jp['Relevancy'] == 'Relevant']
test_jp_filtered = test_jp[test_jp['Relevancy'] == 'Relevant']
```

##### ESG_type

```python
train_jp_filtered['Environment'] = 0
train_jp_filtered['Social'] = 0
train_jp_filtered['Governance'] = 0
train_jp_filtered['None'] = 0

cols = ['Environment', 'Social', 'Governance', 'None']
train_jp_filtered[cols] = train_jp_filtered[cols].astype(float)

train_jp_filtered.loc[train_jp_filtered['ESG_type'] =='Environment', 'Environment'] = 1
train_jp_filtered.loc[train_jp_filtered['ESG_type'] =='Social_Capital','Social'] = 1
train_jp_filtered.loc[train_jp_filtered['ESG_type'] =='Governance','Governance'] = 1
train_jp_filtered.loc[train_jp_filtered['ESG_type'] =='', 'None'] = 1
train_jp_filtered.loc[train_jp_filtered['ESG_type'] =='ESG_general',['Environment', 'Social', 'Governance', 'None']] = [0.3333, 0.3333, 0.3333, 0]

train_jp_filtered = train_jp_filtered[train_jp_filtered['ESG_type'] != 'Human_Capital']
```

```python
test_jp_filtered['Environment'] = 0.0
test_jp_filtered['Social'] = 0.0
test_jp_filtered['Governance'] = 0.0
test_jp_filtered['None'] = 0.0

cols = ['Environment', 'Social', 'Governance', 'None']
test_jp_filtered[cols] = test_jp_filtered[cols].astype(float)

test_jp_filtered.loc[test_jp_filtered['ESG_type'] =='Environment', 'Environment'] = 1
test_jp_filtered.loc[test_jp_filtered['ESG_type'] =='Social_Capital','Social'] = 1
test_jp_filtered.loc[test_jp_filtered['ESG_type'] =='Governance','Governance'] = 1
test_jp_filtered.loc[test_jp_filtered['ESG_type'] =='', 'None'] = 1
test_jp_filtered.loc[test_jp_filtered['ESG_type'] =='ESG_general',['Environment', 'Social', 'Governance', 'None']] = [0.3333, 0.3333, 0.3333, 0]

test_jp_filtered = test_jp_filtered[test_jp_filtered['ESG_type'] != 'Human_Capital']
```

##### Impact_type

```python
train_jp_filtered['pos'] = 0.0
train_jp_filtered['neu'] = 0.0
train_jp_filtered['neg'] = 0.0

train_jp_filtered.loc[train_jp_filtered['Impact_type'] == 'Opportunity', 'pos'] = 1
train_jp_filtered.loc[train_jp_filtered['Impact_type'] == 'Risk', 'neg'] = 1
train_jp_filtered.loc[train_jp_filtered['Impact_type'] == '', 'neu'] = 1
```

```python
test_jp_filtered['pos'] = 0.0
test_jp_filtered['neu'] = 0.0
test_jp_filtered['neg'] = 0.0

test_jp_filtered.loc[test_jp_filtered['Impact_type'] == 'Opportunity', 'pos'] = 1
test_jp_filtered.loc[test_jp_filtered['Impact_type'] == 'Risk', 'neg'] = 1
test_jp_filtered.loc[test_jp_filtered['Impact_type'] == '', 'neu'] = 1
```

Clean the unecessery attributes
- Relevancy
- ESG_type
- Impact_type

```python
train_jp_filtered = train_jp_filtered.drop(columns=['Relevancy', 'ESG_type', 'Impact_type'])
test_jp_filtered = test_jp_filtered.drop(columns=['Relevancy', 'ESG_type', 'Impact_type'])
```

```python
train_jp_filtered.to_csv("datasets/update_datasets/other_languages/train_jp_to_en_filtered.csv", index=False)
test_jp_filtered.to_csv("datasets/update_datasets/other_languages/test_jp_to_en_fitered.csv", index=False)
```

### Mixing French dataset and English dataset 
An attempt will be made to merge the French and English datasets in order to construct a more balanced dataset. It is well established that machine learning models generally perform more effectively and achieve higher predictive accuracy when trained on balanced datasets.

```python
train_df = pd.read_csv('datasets/update_datasets/train_df.csv')
train_fr = pd.read_csv("datasets/update_datasets/other_languages/train_fr_to_en.csv")
train_df = train_df.drop(columns = ['combination'])
train_fr["impact_level"] = train_fr["impact_level"].replace(["low","medium","high"],[0,1,2])
train_fr["impact_length"] = train_fr["impact_length"].replace(["Less than 2 years","2 to 5 years","More than 5 years"],[0,1,2])
train_fr_en = pd.concat([train_df, train_fr], ignore_index=True)
```

```python
train_df
```

```python
train_fr_en
```

```python
balanced_df = data_enrichment.balance_two_columns_by_undersample(
    df=train_fr_en,
    col_a='impact_level',
    col_b='impact_length',
    strategy='median',   
    random_state=123,
    verbose=True
)
```

```python
balanced_df.to_csv("datasets/update_datasets/other_languages/train_balanced.csv", index=False)
```
