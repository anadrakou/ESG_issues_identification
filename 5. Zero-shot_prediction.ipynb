### Zero Shot prediction - GPT 3.5

GPT-3.5 was utilized to predict both the impact level and impact length. A zero-shot prediction approach was manually applied, and the results were stored in a JSON file. The evaluation will focus on calculating the F1-scores, without including the AUC-PR metric, as it cannot be derived in this context. This limitation arises because the AUC-PR requires probabilistic outputs or decision scores across different thresholds, whereas the predictions from GPT-3.5 are provided only as discrete class labels.

```python
from sklearn.metrics import f1_score
import pandas as pd
```

```python
test_true = pd.read_json('datasets/competition/ML-ESG3_Testset_EN.json')
test_predicted = pd.read_json('GPT_results.json')

test_true_level = test_true[['impact_level']]
test_predicted_level = test_predicted[['impact_level']]
test_true_length = test_true[['impact_length']]
test_predicted_length = test_predicted[['impact_length']]

```

```python
f1 = f1_score(test_true_level, test_predicted_level, average='macro')  
print(f"F1 Score for the impact level is : {f1}")
```

```python
f1 = f1_score(test_true_length, test_predicted_length, average='macro')  
print(f"F1 Score for the impact length is : {f1}")
```
